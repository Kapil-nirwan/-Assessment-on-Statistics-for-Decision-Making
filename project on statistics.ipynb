# Step 1: Handle Missing Values
# Droping rows where 'Price' is missing

```python

df_cleaned = df.dropna(subset=['Price'])
df
```

# Filling missing values in 'Car', 'BuildingArea', and 'YearBuilt' with median values

```python
df_cleaned['Car'].fillna(df_cleaned['Car'].median(), inplace=True)
df_cleaned['BuildingArea'].fillna(df_cleaned['BuildingArea'].median(), inplace=True)
df_cleaned['YearBuilt'].fillna(df_cleaned['YearBuilt'].median(), inplace=True)


# Filling missing 'CouncilArea' with the most frequent value

df_cleaned['CouncilArea'].fillna(df_cleaned['CouncilArea'].mode()[0], inplace=True)
df
```

# Step 2: Handle Duplicates


```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
df_cleaned.head()
```

# Step 3: Remove Outliers in 'Price'
# Using IQR to remove outliers in 'Price'


```python
Q1 = df_cleaned['Price'].quantile(0.25)
Q3 = df_cleaned['Price'].quantile(0.75)
IQR = Q3 - Q3
df_cleaned = df_cleaned[(df_cleaned['Price'] >= (Q1 - 1.5 * IQR)) & (df_cleaned['Price'] <= (Q3 + 1.5 * IQR))]
```

# Q1. For the suburb Altona, it is postulated that a typical property sells for $800,000. Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? Use a significance level of 5%. 

# Step 1: Filtering the data for the suburb 'Altona'

```python

altona_data = df[df['Suburb'] == 'Altona']

# Step 2: Performing a one-sample t-test
# Hypothesized population mean price
hypothesized_price = 800000

# Removing missing values in 'Price'
altona_prices = altona_data['Price'].dropna()

# Performing the one-sample t-test
t_stat, p_value = stats.ttest_1samp(altona_prices, hypothesized_price)

# Step 3: Results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
# Step 4: Comparing with significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The typical property price is significantly different from $800,000.")
else:
    print("Fail to reject the null hypothesis. The typical property price is not significantly different from $800,000.")
```

# Step 1: Ensureing the 'Date' column is in datetime format to avoid .dt accessor issues

```python
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce', format='%d/%m/%Y')
```

# Step 2: Filter data for the year 2016


```python
df_2016 = df_cleaned[df_cleaned['Date'].dt.year == 2016]
```

# Defining winter (October to March) and summer (April to September) months


```python
winter_months = [10, 11, 12, 1, 2, 3]
summer_months = [4, 5, 6, 7, 8, 9]
```

# Separating the data into winter and summer months

```python

winter_data = df_2016[df_2016['Date'].dt.month.isin(winter_months)]
summer_data = df_2016[df_2016['Date'].dt.month.isin(summer_months)]
```


# Removing missing values in 'Price' for both groups

```python
winter_prices = winter_data['Price'].dropna()
summer_prices = summer_data['Price'].dropna()
```

# Perform an independent t-test to compare means

```python
t_stat, p_value = stats.ttest_ind(winter_prices, summer_prices)
```

# Step 3: Print the test results

```python
print(f"\nT-statistic (Summer vs Winter 2016): {t_stat}")
print(f"P-value (Summer vs Winter 2016): {p_value}")
```

# Step 4: Compare with significance level


```python
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in property prices between summer and winter months in 2016.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in property prices between summer and winter months in 2016.")
```

# Q3

# Step 1: Filtering data for the suburb 'Abbotsford'

```python
abbotsford_data = df_cleaned[df_cleaned['Suburb'] == 'Abbotsford']
```

# Step 2: Calculating the probability of no car parking

```python
# Assuming 'Car' column indicates the number of parking spaces
# We'll assume properties with 'Car' value of 0 represent no car parking
no_parking_count = (abbotsford_data['Car'] == 0).sum()
total_properties = len(abbotsford_data)
```

# Probability of a property without car parking (p)

```python
p_no_parking = no_parking_count / total_properties
```

```python
p_no_parking
```

# Step 3: Calculate the probability using binomial distribution

```python
n = 10  # Total number of properties
k = 3   # Number of properties without parking
```

# Binomial probability of exactly 3 properties without parking out of 10

```python
probability = binom.pmf(k, n, p_no_parking)
```

# Step 4: Print the result rounded to 3 decimal places

```python
print(f"The probability that 3 out of 10 properties will not have car parking is: {round(probability, 3)}")
```
